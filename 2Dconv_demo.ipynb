```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import rfft2,irfft2
%matplotlib inline
```

```python
Ns=32*3
## clearly, beam must be centered around zero!
x=np.outer(np.ones(Ns),np.hstack((arange(0,Ns/2),arange(-Ns/2,0))))
y=x.T
beam=(exp(-x**2/(2*5**2)-y**2/(2*10**2)))
beam/=beam.sum()
plt.imshow(beam)
```

```python
mp=np.zeros((Ns,Ns))
mp[((x-40)**2+(y-40)**2)<10000]=1
plt.imshow(mp,interpolation='nearest')
plt.colorbar()
```

```python
conv=irfft2(rfft2(beam)*rfft2(mp))
plt.imshow(conv)
plt.colorbar()
```
